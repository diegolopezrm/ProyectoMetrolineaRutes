# Definicion de las funciones Checker2 y Checker
* La funcion ***checker*** tiene la misma funcionalidad de evaluar una cadena que se le ingrese al AFD
* La funcion ***checker2*** tiene la funcionalidad de evaluar una cadena que se le ingrese al AFD para posteriormente traducirla a un lenguaje entendible al usuario de metrolinea

```python
def translate_es(name):
    return name_estaciones[name]
def translate_bus(name1):
    return name_bus[name1]
```

```python
def checker2(cadena,DFA):
    print([step for step in DFA.validate_input(cadena, step=True)])
    print("La secuencia de buses son:\n")
    est = [translate_es(step) for step in DFA.validate_input(cadena, step=True)]
    nn = 0
    for c in cadena:
        d= nn+1
        print(est[nn]," > ",translate_bus(c),' > ',est[d])
        nn = nn+1
   
    
def checker(cadena,DFA):
        try:
            DFA.validate_input(cadena)
            return checker2(cadena,DFA)
        except Exception as e:
            print("Entrada no valida: \n", e)
```
